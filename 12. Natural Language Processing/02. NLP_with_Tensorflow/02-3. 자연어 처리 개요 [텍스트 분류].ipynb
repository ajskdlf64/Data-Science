# 텍스트 분류

텍스트 분류(Text Classifitcation)는 자연어 처리 문제 중 가장 대표적이고 많이 접하는 문제이다. 자연어 처리 기술을 활용해 특정 텍스트를 사람들이 정한 몇 가지 범주(class) 중 어느 범주에 속하는지 분류하는 문제이다. 보통 2가지 범주에 대해 구분하는 문제를 이진 분류(Binary Classification) 문제라 하고, 3가지 이상의 범주에 대해 분류하는 문제를 통틀어서 다중 범주 분류(Multi class Classification) 문제라 한다.
