# 자연어 생성

언어 생성이란 무엇일까? 사람이 언어를 생성하는 과정을 생각해보자. 일반적으로 글을 쓰거나 말을 할 때 어떠한 주제에 대한 목적 의식을 가지고 언어제 맞는 문법과 올바른 단어를 사용해 문장을 생성한다. 신문 기사가 될 수도 있고, 상대방과의 대화, 문장 요약 등 언어를 활용해 우리는 서로 소통(Communication)하면서 살아간다.

그렇다면 컴퓨터가 상대방의 대화를 이해하고, 글도 쓴다면 어떨까? 로봇에게만 적용해도 우리의 삶은 현재와는 다른 세상이 될 것이다. 실제로 기술이 진보하면서 이와 관련된 사례들이 조금씩 나오고 있다.

미국 캘리포니아 LA Times 에서는 "Quakebot"이 인공지능으로 지진이 일어난지 3분만에 기사를 작성했다. 기사의 내용을 보면 지진이 난 경과와 날짜, 일어난 위치를 정확하게 설명하고 있고, 맨 마지막에는 이 기사가 인공지능 알고리즘으로 작성한 기사라고 쓰여있다.

하지만 이런 사실 기반의 기사를 제외하고는 일반적으로 감정 및 논리력이 들어가는 기사를 작성하는 데에는 아직까지 한계가 있어서 보조 도구로 쓰이고 있는 상태이며, 관련 영역에 해당하는 많은 데이터가 필요하다.

컴퓨터는 0과 1로 이루어져 있다. 따라서 감정이나 논리 같은 숫자로 정량하하기 어려운 내용이 포함된 문장에 대해서는 아직까지는 컴퓨터가 작성하기 어렵지만 사실 기반의 기록에는 충분히 효과적이다. 예를 들어 비즈니스에 관한 분석한 결과를 글로 표현하거나, 이메일이나 문자의 내용을 보고 자주 쓰는 답변으로 답하거나, 일반 가전기기가 고장났을 때 고장 내용을 작성해 주는 등 활용할 가치가 많다.
